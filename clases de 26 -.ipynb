# 27. Leer los datos desde una URL externa.

```python
medals_url='http://winterolympicsmedals.com/medals.csv'
```

```python
medals_data=pd.read_csv(medals_url)
```

```python
medals_data.head()
```

* decode --> decodificar
* encode --> codificar

```python
import urllib3

http = urllib3.PoolManager()
resp = http.request("GET", medals_url)
resp_binarie=resp.data #en binario-string
resp_string=resp_binarie.decode('ascii')

print(resp_string)
```

```python
response = resp.data
import csv
cr=csv.reader(resp_string)
for row in cr:
    print(row)
```
